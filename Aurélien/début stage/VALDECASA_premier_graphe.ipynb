```python
import pandas as pd

df = pd.read_csv("op and ch.csv")
df
```

```python
df = df[["Row Key", "Col Key", "Count"]]
df
```

```python
df = df.dropna(subset=['Row Key', 'Col Key'])
```

```python
df
```

```python
def create_gephi(df):
    edges = []
    nodes = set()
    
    for _, row in df.iterrows():
        attr_words = row['Row Key'].split(', ')
        obj_words = row['Col Key'].split(', ')
        weight = row['Count']
        
        for attr in attr_words:
            for obj in obj_words:
                edges.append((attr, obj, weight))
                edges.append((obj, attr, weight))
                nodes.add(attr)
                nodes.add(obj)
    
    # Sauvegarde des nœuds
    nodes_df = pd.DataFrame({'Id': list(nodes)})
    
    # Sauvegarde des arêtes
    edges_df = pd.DataFrame(edges, columns=['Source', 'Target', 'Weight'])
    
    display(nodes_df)
    display(edges_df)

    edges_df.to_csv("edges.csv", index=False)
    nodes_df.to_csv("nodes.csv", index=False)

    return(nodes_df, edges_df)

# Exemple d'utilisation
create_gephi(df)
```
