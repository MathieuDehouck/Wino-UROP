```python
df = pd.read_json("C:/Users/valde/Documents/GitHub/Wino-UROP/lemma_train_xl.jsonl", lines=True)
df
```

### filtrage df

```python
# Supprimer les lignes où la colonne "answer" contient la valeur "2"
df = df[df['answer'] != 2]

# Sélectionner les colonnes "lemma_option1" et "lemma_option2"
df = df[['lemma_option1', 'lemma_option2']]
df
```

```python
edge_counts = df.groupby(['lemma_option1', 'lemma_option2']).size().reset_index(name='Weight')
edge_counts = edge_counts.sort_values(by='Weight', ascending=False)
edge_counts.columns = ['Source', 'Target', 'Weight']
edge_counts.to_csv('edges.csv', index=False)
display(edge_counts)
```

```python
df = edge_counts

# Step 1: Extract unique nodes
unique_nodes = pd.unique(df[['Source', 'Target']].values.ravel('K'))

# Step 2: Initialize the adjacency matrix
adj_matrix = pd.DataFrame(0, index=unique_nodes, columns=unique_nodes)

# Step 3: Populate the adjacency matrix
for _, row in df.iterrows():
    adj_matrix.at[row['Source'], row['Target']] = row['Weight']

# Display the adjacency matrix
print(adj_matrix)
```
