### filtrage df

```python
# Supposons que vous voulez afficher la ligne où groupID est '3QHITW7OYO7Q6B6ISU2UMJB84ZLAQE-1'
group_id = '3D5G8J4N5CI2K40F4RZLF9OG2CKVTH-2'
specific_row = df.loc[df['qID'] == group_id, "lemma_common_tokens"]

specific_list = specific_row.tolist()
print(specific_list)
```

```python
# Fonction pour nettoyer les tokens
def clean_tokens(tokens):
    # Filtrer les tokens pour enlever les ponctuations et les underscores
    return [token for token in tokens if token not in string.punctuation and token != '_']

# Appliquer la fonction à la colonne 'lemma_common_tokens'
df['lemma_common_tokens'] = df['lemma_common_tokens'].apply(clean_tokens)

# Fonction pour ajouter les options à la liste de tokens
def add_options_to_tokens(row):
    return row['lemma_common_tokens'] + [row['lemma_option1'], row['lemma_option2']]

# Appliquer la fonction à chaque ligne du DataFrame
df['lemma_common_tokens'] = df.apply(add_options_to_tokens, axis=1)

df = df[["lemma_common_tokens","lemma_special_alternate"]]

df = df.rename(columns={
    'lemma_special_alternate': 'SA',
    'lemma_common_tokens': 'mots'
})

df
```

```python
# Initialiser une liste pour stocker les arêtes
edges = []

# Parcourir chaque ligne du DataFrame
for index, row in df.iterrows():
    # Pour chaque élément dans SA
    for sa in row['SA']:
        # Créer une connexion avec chaque mot dans mots
        for mot in row['mots']:
            edges.append({'Source': sa, 'Target': mot})

# Créer un DataFrame à partir de la liste des arêtes
edges_df = pd.DataFrame(edges)

# Compter les occurrences de chaque paire unique et ajouter une colonne de poids
edges_df['Weight'] = edges_df.groupby(['Source', 'Target'])['Source'].transform('count')

# Garder uniquement les paires uniques
edges_df = edges_df.drop_duplicates()

display(edges_df)
```

```python
# Enregistrer le DataFrame au format CSV
edges_df.to_csv('edges.csv', index=False)
```
