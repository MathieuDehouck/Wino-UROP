```python
df = pd.read_json("C:/Users/valde/Documents/GitHub/Wino-UROP/train_xl.jsonl", lines=True)
df
```

```python
# Liste des mots Ã  rechercher
mots = ['empty', 'full', 'box', 'bag', 'bowl', 'bottle', 'cup', 'pan', 'pot']

# Fonction pour vÃ©rifier si une ligne contient au moins un des mots
def contient_mots(row, mots):
    return any(mot in row['sentence'] or mot in row['option1'] or mot in row['option2'] for mot in mots)

# Filtrer les lignes
df_filtre = df[df.apply(contient_mots, axis=1, mots=mots)]

# Afficher le DataFrame filtrÃ©
df_filtre
```
