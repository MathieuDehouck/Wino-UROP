```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("classe_nourriture_entier.csv")
df = df.drop('Unnamed: 0', axis=1)

df
```

```python
class_counts = df['classe'].value_counts()

plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar', color='teal')

plt.title('Répartition des classes')
plt.xlabel('Classes')
plt.ylabel('Fréquence')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)


plt.tight_layout()
plt.show()
```

```python
class_counts
```

```python
plt.figure(figsize=(12, 6))

plt.hist(df['probabilite_nourriture'], 
         bins=50, 
         color='teal', 
         edgecolor='black', 
         alpha=0.7)

plt.title('Distribution des probabilités de "nourriture"')
plt.xlabel('Probabilité')
plt.ylabel('Fréquence')
plt.grid(axis='y', alpha=0.4)
plt.xlim(0, 1) 
```

```python
# Top 10 des probabilités les plus élevées
top10 = df.nlargest(10, 'probabilite_nourriture').assign(position='Top')

# Top 10 des probabilités les plus basses
bottom10 = df.nsmallest(10, 'probabilite_nourriture').assign(position='Bottom')
```
