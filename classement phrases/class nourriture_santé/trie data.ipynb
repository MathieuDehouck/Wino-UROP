```python
import pandas as pd
```

```python
df = pd.read_json("train_xl.jsonl", lines=True)
df = df[df['answer'] == 1]
df.rename(columns={'sentence': 'phrase'}, inplace=True)
pd.set_option('display.max_colwidth', None)

df
```

```python
# Liste des mots-clés à rechercher
mots_cles = [
    # Food-related nouns
    "fruit", "vegetable", "protein", "carbohydrate", "fat", "fiber",
    "snack", "meal", "calorie", "diet", "nutrition", "ingredient",
    "portion", "superfood","candy","cake","food", "sugar", "weight","sweeter", "sweet", "health", "wellness", "weight", "metabolism",

    # Food-related adjectives
    "nutritious", "organic", "low-fat", "sugar-free",
     "fresh",

    # Food-related verbs (conjugated)
    "eat", "eats", "ate", "eaten", "eating",
    "cook", "cooks", "cooked", "cooking",
    "serve", "serves", "served", "serving",


    # Health/slimness-related adjectives
    "healthy", "slim", "lean",
    "balanced", "overweight", "underweight",

    # Health/slimness-related verbs (conjugated)
    "burn", "burns", "burned", "burning",
    "gain", "gains", "gained", "gaining",
    "stretch", "stretches", "stretched", "stretching"
]
# Créer une expression régulière pour matcher les mots entiers
pattern = r'\b(?:{})\b'.format('|'.join(mots_cles))

# Filtrer les lignes qui contiennent au moins un des mots-clés
df = df[df['phrase'].str.contains(pattern, case=False, na=False)]

df
```
