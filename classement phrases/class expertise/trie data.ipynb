```python
import pandas as pd
```

```python
df = pd.read_json("train_xl.jsonl", lines=True)
df = df[df['answer'] == 1]
df.rename(columns={'sentence': 'phrase'}, inplace=True)
pd.set_option('display.max_colwidth', None)

df
```

```python
# Liste des mots-clÃ©s Ã  rechercher
mots_cles = [
    # Adjectifs
    "skilled",
    "experienced",
    "competent",
    "capable",
    "qualified",
    "efficient",
    "knowledgeable",
    "talented",
    "gifted",
    "versatile",

    # Adverbes / comparateurs
    "better",
    "quicker",
    "faster",
    "more effectively",
    "more accurately",
    "more precisely",
    "more reliably",
    "more creatively",
    "more efficiently",

    # Expressions idiomatiques
    "ahead of the curve",
    "a step ahead",
    "stands out",
    "outperforms",
    "sets the standard",
    "has the edge",
    "leads the pack",
    "brings more to the table"
]
# CrÃ©er une expression rÃ©guliÃ¨re pour matcher les mots entiers
pattern = r'\b(?:{})\b'.format('|'.join(mots_cles))

# Filtrer les lignes qui contiennent au moins un des mots-clÃ©s
df = df[df['phrase'].str.contains(pattern, case=False, na=False)]

df
```

```python
df.to_csv('phrases_annotees.csv', index=False)
```
