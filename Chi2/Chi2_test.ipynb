```python
df = pd.read_csv("label_genre_nourriture.csv")
df
```

```python
df['label'] = df['probabilite_nourriture'].apply(lambda x: 'pos' if x > 0.5 else 'neg')
df = df[df['genre'] != 'f_h']
```

```python
# Créer un tableau de contingence
contingency_table = pd.crosstab(df['genre'], df['label'])

# Effectuer le test du Chi2
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Tableau de contingence:")
print(contingency_table)
print("\nStatistique Chi2:", chi2)
print("Valeur p:", p)
```
